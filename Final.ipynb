```python
import pandas as pd
```

```python
food = pd.read_csv('Data/Food.csv',)
Grocery= pd.read_csv('Data/Grocery.csv')
Medicals = pd.read_csv('Data/Medicals.csv')
Shooping = pd.read_csv('Data/Shooping.csv')
```

```python
datamapping= {
    "Food" : 1,
    "Grocery" : 2,
    "Medical" : 3,
    "Shopping" : 4
}
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
targets = df['Target']
```

```python
final=np.array([targets])
```

```python
final
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

col = final

# Standardize the data
scaler = StandardScaler()
col = scaler.fit_transform(col.reshape(-1, 1))

# Fit the KMeans model
kmeans = KMeans(n_clusters=4)
kmeans.fit(col)

# Predict the cluster labels
labels = kmeans.predict(col)

# Compute the silhouette score
score = silhouette_score(col, labels)
print("Silhouette Score:", score)
```
